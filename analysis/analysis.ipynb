## Dados brutos

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./processed_data.csv')
```

```python
# Plotando os dados brutos
plt.figure(figsize=(10, 6))

plt.plot(df["Processos"], df["16 tasks"], label="16 tasks", marker="o")
plt.plot(df["Processos"], df["32 tasks"], label="32 tasks", marker="s")

plt.title("Tempo de Execução vs. Quantidade de Processos")
plt.xlabel("Quantidade de Processos")
plt.ylabel("Tempo de Execução (segundos)")
plt.legend()
plt.grid(True)

plt.show()
```

## Cálculo de Speedup 

```python
# Speedup
df["Speedup (16 tasks)"] = max(df["16 tasks"]) / df["16 tasks"]
df["Speedup (32 tasks)"] = max(df["32 tasks"]) / df["32 tasks"]

plt.figure(figsize=(10, 6))
plt.plot(df["Processos"], df["Speedup (16 tasks)"], label="16 tasks", marker="o")
plt.plot(df["Processos"], df["Speedup (32 tasks)"], label="32 tasks", marker="s")


plt.title("Speedup 16 tasks x 32 tasks")
plt.xlabel("Quantidade de Processos")
plt.ylabel("Speedup")
plt.legend()
plt.grid(True)

plt.show()
```

## Cálculo de Eficiência

```python
# Eficiência
df["Eficiência (16 tasks)"] = df["Speedup (16 tasks)"] / df["Processos"]
df["Eficiência (32 tasks)"] = df["Speedup (32 tasks)"] / df["Processos"]


plt.figure(figsize=(10, 6))
plt.plot(df["Processos"], df["Eficiência (16 tasks)"], label="16 tasks", marker="o")
plt.plot(df["Processos"], df["Eficiência (32 tasks)"], label="32 tasks", marker="s")


plt.title("Eficiência 16 tasks x 32 tasks")
plt.xlabel("Quantidade de Processos")
plt.ylabel("Eficiência")
plt.legend()
plt.grid(True)


plt.show()
```
